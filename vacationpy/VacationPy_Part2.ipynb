### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
city_df = pd.read_csv("weather_city_data.csv")


ideal_weather=city_df.loc[(city_df["Max Temp"]<80) & 
                          (city_df["Max Temp"]>70) & 
                          (city_df["Wind Speed"]<10) & 
                          (city_df["Cloudiness"]==0), :]

print(f'Total number of cities that match these criteria: {len(ideal_weather)}')
      

      
hotel_df=ideal_weather[["City", "Country", "Lat", "Lng"]]
hotel_df["Hotel Name"]=""
hotel_df["Hotel City"]=""
hotel_df["Hotel Country"]=""
hotel_df
```

```python
import json
```

```python
# testing for one city!

base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

lat=hotel_df.iloc[0, 2]
lng=hotel_df.iloc[0, 3]

params = {
    "location": f'{lat},{lng}',  
    "radius": "5000",
    "type": "lodging",
    "key": g_key,
}


                                      
response=requests.get(base_url, params=params).json()
results=response["results"][0]

print(json.dumps(results, indent=4, sort_keys=True))

```

```python
# testing for hotel info in 1 city!

hotel_name=results['name']
hotel_vicinity=results['vicinity']
hotel_lat=results["geometry"]["location"]["lat"]
hotel_lng=results["geometry"]["location"]["lng"]


print(hotel_name)
print(hotel_vicinity)
print(hotel_lat)
print(hotel_lng)
```

```python
# Homework: hotel_df run! (not testing!)

hotel2_df=hotel_df.reset_index(drop=True)

hotel2_df["Hotel Vicinity"]=""
hotel2_df["Hotel Latitude"]=""
hotel2_df["Hotel Longitude"]=""



for index, row in hotel2_df.iterrows():
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
   
    lat=hotel_df.iloc[index, 2]
    lng=hotel_df.iloc[index, 3]    
    
    loc=f'{lat},{lng}'
    
    city=row["City"]
    print(f"Retrieving results for Index {index} in {city}.")
#     print(f"The geolocation for {city} is: {loc}")
    
    params = {
    "location": loc,  
    "radius": "5000",
    "type": "lodging",
    "key": g_key,
    }
                                          
    response=requests.get(base_url, params=params).json()
    results=response["results"][0]
    
    hotel2_df.iloc[index, 4]=results['name']
    hotel2_df.iloc[index, 7]=results['vicinity']
    hotel2_df.iloc[index, 8]=results["geometry"]["location"]["lat"]
    hotel2_df.iloc[index, 9]=results["geometry"]["location"]["lng"]

    
```

```python
for index, row in hotel2_df.iterrows():
    
    hotel_lat=hotel2_df.iloc[index, 8]
    hotel_lng=hotel2_df.iloc[index, 9]
    
    loc=f'{hotel_lat},{hotel_lng}'
    
    base_url= ("https://maps.googleapis.com/maps/api/geocode/json?latlng={0}&key={1}").format(loc, g_key)
    
    city=row["City"]
    print(f"Retrieving results for Index {index} in {city}.")
                                          
    response=requests.get(base_url).json()
    results=response["results"][0]
    
    try:
        
        hotel2_df.iloc[index, 5]=row["City"]
        hotel2_df.iloc[index, 6]=results["address_components"][5]["long_name"]
    except:
        hotel2_df.iloc[index, 5]=row["City"]
        hotel2_df.iloc[index, 6]=results["address_components"][3]["long_name"]
    
    
    
    
```

```python
# Create the heat map for hotel2_df

# from config import gkey

gmaps.configure(api_key=g_key)

# Store latitude and longitude in locations
geoloc = hotel2_df[["Hotel Latitude", "Hotel Longitude"]]



# Plot Heatmap
fig = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(geoloc)


# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel2_df.iterrows()]
marker_locations = hotel2_df[['Hotel Latitude', 'Hotel Longitude']]
```

```python
# Create a marker_layer using the poverty list to fill the info box

markers = gmaps.marker_layer(marker_locations,
                             info_box_content=hotel_info)
fig.add_layer(markers)

# Display Map
fig
```
